## Imports

```python
import pandas as pd
import kagglehub
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

## Download of the dataset

```python
path = kagglehub.dataset_download("uciml/default-of-credit-card-clients-dataset")
path
```

```python
df = pd.read_csv(path + "/UCI_Credit_Card.csv")
df
```

## Plots of the dataset

```python
plt.hist(df['LIMIT_BAL'], bins=30, color='lightgreen', alpha=0.7)
plt.title('Distribution of Credit Limit Balance')
plt.xlabel('Credit Amount in NT$')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

ax = plt.gca()
ax.ticklabel_format(style='plain', axis='x')

plt.show()
```

```python
plt.hist(df['AGE'], bins=30, color='lightgreen', alpha=0.7)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#plot the gender distribution
gender_counts = df["SEX"].map({1: 'Male', 2: 'Female'}).value_counts()

plt.bar(gender_counts.index, gender_counts.values, color=['tomato', 'cyan'], alpha=0.7)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
